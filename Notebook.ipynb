```python
target = 'value_euro'
y_log = np.log1p(df_model[target]).rename('value_log')
```

```python
candidate_features = [
    # personal / numeric
    'age', 'height_cm', 'weight_kgs',
    # summary ratings
    'overall_rating', 'potential',
    # financial
    'wage_euro',
    # small categorical descriptors (we'll encode later)
    'preferred_foot', 'body_type', 'positions', 'nationality',
    # skill attributes (common FIFA attributes)
    'crossing','finishing','heading_accuracy','short_passing','volleys','dribbling','curve',
    'freekick_accuracy','long_passing','ball_control','acceleration','sprint_speed','agility',
    'reactions','balance','shot_power','jumping','stamina','strength','long_shots','aggression',
    'interceptions','positioning','vision','penalties','composure','marking','standing_tackle',
    'sliding_tackle'
]
len(candidate_features)
```

```python
features = [c for c in candidate_features if c in df_model.columns]
```

```python
X = df_model[features].copy()
y = y_log.copy()
```

```python
numeric_feats = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_feats = [c for c in X.columns if c not in numeric_feats]
```

```python
print("Target (original) column" , target)
print("Created log- targeted columns name : 'value_log'")
print(X.shape)
print(y.shape)
```

```python
from sklearn.pipeline import Pipeline 
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import StandardScaler, OneHotEncoder 
from sklearn.compose import ColumnTransformer
```

```python
numeric_pipeline = Pipeline(steps=[('imputer',SimpleImputer(strategy='median')),
                                   ('scaler', StandardScaler())
                                  ])
```

```python
categorical_pipeline = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
                                       ('onehot' , OneHotEncoder(handle_unknown = 'ignore',sparse_output= False))
                                      ])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num',numeric_pipeline,numeric_feats),
    ('cat',categorical_pipeline,categorical_feats)
], remainder='drop')
preprocessor.fit(X)
```

```python
sample = preprocessor.transform(X.iloc[:5])
sample.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor 
from sklearn.pipeline import Pipeline 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
rf = RandomForestRegressor(n_estimators=200,random_state=42,n_jobs=-1)
```

```python
model_pipeline = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('regressor',rf)
])
```

```python
model_pipeline.fit(X_train,y_train)
```

```python
preds_log = model_pipeline.predict(X_test)
```

```python
y_test_orig = np.expm1(y_test)
preds_orig = np.expm1(preds_log)
```

```python
mae = mean_absolute_error(y_test_orig,preds_orig)
rmse = np.sqrt(mean_squared_error(y_test_orig,preds_orig))
r2 = r2_score(y_test_orig,preds_orig)
```

```python
print(f"\nEvaluation on test set (original euro scale):")
print(f"MAE : €{mae:,.2f}")
print(f"RMSE: €{rmse:,.2f}")
print(f"R²  : {r2:.4f}")
```

```python
mae_log = mean_absolute_error(y_test, preds_log)
rmse_log = np.sqrt(mean_squared_error(y_test, preds_log))
r2_log = r2_score(y_test, preds_log)
print(f"\nEvaluation on log-target space:")
print(f"MAE (log) : {mae_log:.4f}")
print(f"RMSE(log) : {rmse_log:.4f}")
print(f"R² (log)  : {r2_log:.4f}")
```

```python
# Results 
sample_df = X_test.reset_index(drop=True).loc[:9]
sample_actual = y_test_orig.reset_index(drop=True).loc[:9]
sample_pred = preds_orig[:10]
```

```python
display_df = sample_df.copy()
display_df['actual_value_euro']= sample_actual.values
display_df['predicted_value_euro'] = sample_pred
print("\n Sample Prediction (first 10 test rows): ")
display(display_df.head(10))
```
